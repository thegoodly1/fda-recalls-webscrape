### This is where I'll make my function that'll be compatible with the remaining webpages I want.

```python
def getting_recalls(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')
    
    for tr in soup.find('table').find('tbody').find_all('tr'):
            rows = {}
            cells = tr.find_all('td')
            rows['Date'] = cells[0].text.strip()
            rows['Brand Name'] = cells[1].text.strip()
            rows['Product Description'] = cells[2].text.strip()
            rows['Reason/Problem'] = cells[3].text.strip()
            rows['Company'] = cells[4].text.strip()
            table.append(rows)
```

#### Calling the function on every url from 2010-2015

```python
table=[]
for u in all_urls:                                         
    getting_recalls(u)
fda_recalls_all = pd.DataFrame(table)           
```
